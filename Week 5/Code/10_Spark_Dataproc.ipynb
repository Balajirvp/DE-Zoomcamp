# Submitting a job from UI

After setting up the dataproc cluster, run the below command in the terminal to copy the code file to our existing GCS bucket (Ideally, we will have separate buckets for codes and data)

```python
# gsutil cp 09_Spark_SQL.py gs://dtc_data_lake_composed-sun-375018/code/09_Spark_SQL.py  (Update your bucket name accordingly)
```

Inside the cluster, click on submit job and fill the relevant details 

<b>Job type</b> - PySpark \
<b>Main python file</b> - Enter the GCS bucket location for the python file we just copied (gs://dtc_data_lake_composed-sun-375018/code/09_Spark_SQL.py) \
<b>Arguments</b> - 

`--input_green=gs://dtc_data_lake_composed-sun-375018/pq/green/2021/*/` \
`--input_yellow=gs://dtc_data_lake_composed-sun-375018/pq/yellow/2021/*/`  
`--output=gs://dtc_data_lake_composed-sun-375018/report-2021`

(Change the bucket name accordingly)

After entering these. click <b>Submit</b>



You can see that the `report-2021` folder is created in the GCS bucket -
